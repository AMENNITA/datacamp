# Plotting multiple graphs

## Strategies
- Plo!ing many graphs on common axes
- Creating axes within a figure
- Creating subplots within a figure


## Graphs on common axes


Assume we have 3 numpy arrays:
- `t`: for time
- `temperature`
- `dewpoint`

Data is from weather measurment from 2 weeks in Jan 2010

```python
In [1]: import matplotlib.pyplot as plt 
In [2]: plt.plot(t, temperature, 'red')
In [3]: plt.plot(t, dewpoint, 'blue') # Appears on same axes
In [4]: plt.xlabel('Date')
In [5]: plt.title('Temperature & Dew Point')
In [6]: plt.show() # Renders plot objects to screen
```
Graph has two curves overlay on the same axes


- overlaying curves is not usefull when the scales differ wildly

## Using `axes()`

The axes command create coordinate axis with a figure within subsequent plots are drawn.


```python
In [1]: plt.axes([0.05,0.05,0.425,0.9]) # axes lefthand 
In [2]: plt.plot(t, temperature, 'red')
In [3]: plt.xlabel('Date')
In [4]: plt.title('Temperature')
In [5]: plt.axes([0.525,0.05,0.425,0.9])  # new axes right fig n active
In [6]: plt.plot(t, dewpoint, 'blue')
In [7]: plt.xlabel('Date')
In [8]: plt.title('Dew Point')
In [9]: plt.show()
    ```


## The `axes()` command

requires:
- the lower left corner
- width
- height of the axes
All express in figure units(numbers between 0 to 1 scale to the dimentions of the figure)

- requires some trial n error

- syntax: `axes([x_lo, y_lo,width,height])`
- Units between 0 and 1 (figure dimentions


## Using `subplot()`

- create a grid of axis (no coordinates manualy)




```python
In [1]: plt.subplot(2, 1, 1) #top
In [2]: plt.plot(t, temperature, 'red')#top
In [3]: plt.xlabel('Date') #top
In [4]: plt.title('Temperature') #top
In [5]: plt.subplot(2, 1, 2) # bott
In [6]: plt.plot(t, dewpoint, 'blue') # bott
In [7]: plt.xlabel('Date') # bott
In [8]: plt.title('Dew Point') # bott
In [9]: plt.tight_layout() # pad spaces between plots
In [10]: plt.show()
    ```
    
## The `subplot()` command
The three subplot arguments are:
- number of rows
- number of col
- wich subplot number to activate

Subplot numbers increase across rows, then down columns.
Starting from `1` on the top left coner.


- Syntax
    - `subplot(nrows, ncols, nsubplot)`
- Subplot ordering:
    - Row-wise from top left
    - indexed from 1

````python
# EXAMPLE 1.2

# Create plot axes for the first line plot
plt.axes([0.05,0.05,0.425, 0.9])

# Plot in blue the % of degrees awarded to women in the Physical Sciences
plt.plot(year, physical_sciences, color='blue')

# Create plot axes for the second line plot
plt.axes([0.525,0.05,0.425, 0.9])


# Plot in red the % of degrees awarded to women in Computer Science
plt.plot(year, computer_science, color='red')

# Display the plot
plt.show()
````

````python
# EXAMPLE 1.3

# Create a figure with 1x2 subplot and make the left subplot active
plt.subplot(1,2,1)

# Plot in blue the % of degrees awarded to women in the Physical Sciences
plt.plot(year, physical_sciences, color='blue')
plt.title('Physical Sciences')

# Make the right subplot active in the current 1x2 subplot grid
plt.subplot(1,2,2)

# Plot in red the % of degrees awarded to women in Computer Science
plt.plot(year, computer_science, color='red')
plt.title('Computer Science')

# Use plt.tight_layout() to improve the spacing between subplots
plt.tight_layout()
#plt.show()
````
