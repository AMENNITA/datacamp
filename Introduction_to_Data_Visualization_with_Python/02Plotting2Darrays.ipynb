## Working with 2D arrays



## Reminder: numpy arrays
- homogeneous in type
- calcualtions all at once
- indexing with brackets:
    - `A[index]` for 1D array
    - `A[index0, index1]` for 2D array
    
## reminder: slicing arrays

- slicing:
    - 1D arrays :`A[slice]`
    - 2D arrays : `A[slice0, slice1]`
- slice = `start:stop:stride`
    - Indexes from `start` to `stop -1` in steps of `stride`
    - Missing start: implicitly at *beggining* of array
    - Missiing stop: implicitly at *end* of array
- Negative indexes/slices: count from *end of array*

## 2D arrays & images


## 2D arrays & functions

```python
## Using `meshgdrid()`

import numpy as np
u = np.linspace(-2,2,3)
v = np.linspace(-1,1,5)

X,Y = np.meshgrid(u,v)

Z = X**2/25 + Y**2/4

## 
print('X:\n{}\n\nY:\n{}'.format(X,Y))    
```

```python
# sampling on a grid
import matplotlib.pyplot as plt
#%matplotlib
plt.set_cmap('gray')
plt.pcolor(Z)
plt.show()
```

# Visualizing bivriate functions

##  Bivariate functions

$f(x,y)= \frac{x^{2}}{25}+\frac{y^{2}}{4}$


## Pseudocolor plot

```python
import numpy as np
import matplotlib.pyplot as plt

u = np.linspace(-2,2,65)
v = np.linspace(-1,1,33)

X,Y = np.meshgrid(u,v)
Z = X**2/25 + Y**2/4

plt.pcolor(Z)
plt.show()
```

## Color bar

```python
# ... continuation from previous code
# ...
plt.pcolor(Z)
plt.colorbar()
plt.show()
```

## Color map

```python
# ... continuation from previous code
# ...
plt.pcolor(Z, cmap='gray')
plt.colorbar()
plt.show()
```

```python
# ... continuation from previous code
# ...
plt.pcolor(Z, cmap='autumn')
plt.colorbar()
plt.show()
```

## Axis tight

```python
# ... continuation from previous code
# ...
plt.pcolor(Z)
plt.colorbar()
plt.axis('tight')

plt.show()
```

## Plot using mesh grid

```python
plt.pcolor(X, Y, Z)  # X,Y,Z are 2D meshgrid
plt.colorbar()
plt.show()

# Axes determined by mesh grid arrays X,Y
```

## Contour plots

```python
plt.contour(Z)
plt.show()
```

# Visualizing bivarite distribution

## Distributions of 2D points

- 2D points given as two 1D arrays *x & y*
- Goal: generate a 2D histogram from *x & y*


## Histogram in 1D
- choose bins(intervals
- count realizations withing bins & plots


## Bins in 2D
- different shapes available from binning points
-  common choices: rectangles & hexagons


```python
import numpy as np
import csv

csvfile = open('auto-mpg.csv','r')

auto = csv.DictReader(csvfile)
# ['mpg', 'cyl', 'displ', 'hp', 'weight', 'accel', 'yr', 'origin', 'name', 'color', 'size', 'marker'
accel = []
weight = []


for row in auto:
    accel.append(float(row['accel']))
    weight.append(float(row['weight']))
    
accel = np.array(accel)
weight = np.array(weight)
```

```python
plt.hist2d(weight,accel, bins=(10,20))  # weight & accel have the same shape (392,)
plt.colorbar()

plt.xlabel('weight ($\mathrm{kg}$)')
plt.ylabel('acceleration ($\mathrm{ms}^{-2}$)')

plt.show()
```

## `Hexbin()`: Hexagonal binning

```python
plt.hexbin(weight,accel, gridsize=(15,10))  # weight & accel have the same shape (392,)
plt.colorbar()
plt.axis('tight')  # axis were showing its hex..!

plt.xlabel('weight ($\mathrm{kg}$)')
plt.ylabel('acceleration ($\mathrm{ms}^{-2}$)')

plt.show()
```

# Working with images

## Images
- grayscale images: rectangular 2D arrays
- Color images: typicall three 2D arrays (channels)
    - RGB (red,green,blue)
- Channel values
    - 0 to 1 ( floating point numbers)
    - 0 to 255 (8bit int)
   

## Loating images

```python
img = plt.imread('google.png')

print(img.shape)
```

```python
plt.imshow(img)

plt.axis('off')

plt.show()
```

## Reduction to gray-scale image

```python
collapsed = img.mean(axis=2)
print(collapsed.shape)
```

```python
plt.set_cmap('gray')
plt.imshow(collapsed, cmap='gray')

plt.axis('off')
plt.show()
```

## Uneven samples

```python
uneven = collapsed[::4,::2]  # nouniform subsampling
print(uneven.shape)  # og was (92, 272, 4)
```

```python
plt.imshow(uneven)

plt.axis('off')
plt.show()
```

## Adjusting aspect ration

```python
plt.imshow(uneven, aspect=2.0)
plt.axis('off')
plt.show()
```

## Adjusting extent

```python
plt.imshow(uneven, cmap='gray', extent=(0,172,0,92))
plt.axis('off')
plt.show()
```
