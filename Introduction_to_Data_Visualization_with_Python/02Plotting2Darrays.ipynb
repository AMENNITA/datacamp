## Working with 2D arrays



## Reminder: numpy arrays
- homogeneous in type
- calcualtions all at once
- indexing with brackets:
    - `A[index]` for 1D array
    - `A[index0, index1]` for 2D array
    
## reminder: slicing arrays

- slicing:
    - 1D arrays :`A[slice]`
    - 2D arrays : `A[slice0, slice1]`
- slice = `start:stop:stride`
    - Indexes from `start` to `stop -1` in steps of `stride`
    - Missing start: implicitly at *beggining* of array
    - Missiing stop: implicitly at *end* of array
- Negative indexes/slices: count from *end of array*

````python
## Using `meshgdrid()`

import numpy as np
u = np.linspace(-2,2,3)
v = np.linspace(-1,1,5)

X,Y = np.meshgrid(u,v)

Z = X**2/25 + Y**2/4

## 
print('X:\n{}\n\nY:\n{}'.format(X,Y))    
````

````python
## Example
# Generating meshes
# Import numpy and matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt


# Generate two 1-D arrays: u, v
u = np.linspace(-2,2, 41)
v = np.linspace(-1,1,21)

# Generate 2-D arrays from u and v: X, Y
X,Y = np.meshgrid(u,v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

# Display the resulting image with pcolor()
plt.pcolor(Z)
plt.show()
````

# Visualizing bivarite distribution

## Distributions of 2D points

- 2D points given as two 1D arrays *x & y*
- Goal: generate a 2D histogram from *x & y*


## Histogram in 1D
- choose bins(intervals
- count realizations withing bins & plots


## Bins in 2D
- different shapes available from binning points
-  common choices: rectangles & hexagons


````python
import numpy as np
import csv

csvfile = open('auto-mpg.csv','r')

auto = csv.DictReader(csvfile)
# ['mpg', 'cyl', 'displ', 'hp', 'weight', 'accel', 'yr', 'origin', 'name', 'color', 'size', 'marker'
accel = []
weight = []
hp = []
mpg = []


for row in auto:
    accel.append(float(row['accel']))
    weight.append(float(row['weight']))
    hp.append(int(row['hp']))
    mpg.append(float(row['mpg']))
    
    
accel = np.array(accel)
weight = np.array(weight)
hp = np.array(hp)
mpg = np.array(mpg)
````

````python
plt.hist2d(weight,accel, bins=(10,20))  # weight & accel have the same shape (392,)
plt.colorbar()

plt.xlabel('weight ($\mathrm{kg}$)')
plt.ylabel('acceleration ($\mathrm{ms}^{-2}$)')

plt.show()
````

## `Hexbin()`: Hexagonal binning

````python
plt.hexbin(weight,accel, gridsize=(15,10))  # weight & accel have the same shape (392,)
plt.colorbar()
plt.axis('tight')  # axis were showing its hex..!

plt.xlabel('weight ($\mathrm{kg}$)')
plt.ylabel('acceleration ($\mathrm{ms}^{-2}$)')

plt.show()
````

````python
## Example
# Using hist2d()
# Generate a 2-D histogram
plt.hist2d(hp,mpg, 
bins=(20,20),
range=((40,235),(8,48)))

# Add a color bar to the histogram
plt.colorbar()

# Add labels, title, and display the plot
plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hist2d() plot')
plt.show()
````

````python
## Example
# Using hexbin()
# http://amstat.tandfonline.com/doi/abs/10.1080/01621459.1987.10478445

# Generate a 2d histogram with hexagonal bins
plt.hexbin(hp,mpg,
            gridsize=(15,12),
            extent=(40,235,8,48))

           
# Add a color bar to the histogram
plt.colorbar()

# Add labels, title, and display the plot
plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hexbin() plot')
plt.show()

````

## Reduction to gray-scale image

````python
collapsed = img.mean(axis=2)  # average the RGB channels by applying np.mean along the axis 2
print(collapsed.shape)
````

````python
plt.set_cmap('gray')
plt.imshow(collapsed, cmap='gray')

plt.axis('off')
plt.show()
````

## Uneven samples

In the grayscale image:
- take the pixels from every 4th row
- take the pizels from every 2nd column

```python
uneven = collapsed[::4,::2]
```
`::4`- every 4th element
`::2`- every 2nd element

````python
uneven = collapsed[::4,::2]  # nouniform subsampling
print(uneven.shape)  # og was (92, 272, 4)
````

````python
plt.imshow(uneven)

plt.axis('off')
plt.show()
````

## Adjusting aspect ration




````python
plt.imshow(uneven, aspect=2.0)
plt.axis('off')
plt.show()
````

## Adjusting extent



````python
plt.imshow(uneven, cmap='gray', extent=(0,172,0,92))
plt.axis('off')
plt.show()
````

````python
## Example
# Pseudocolor plot from image data

# Load the image into an array: img
img = plt.imread('Astronaut.jpg')

# Print the shape of the image
print(img.shape)

# Compute the sum of the red, green and blue channels: intensity
intensity = img.sum(axis=2)

# Print the shape of the intensity
print(intensity.shape)

# Display the intensity with a colormap of 'gray'
plt.imshow(intensity, cmap='gray')

# Add a colorbar
plt.colorbar()

# Hide the axes and show the figure
plt.axis('off')
plt.show()
````

````python
## Example
#Extent and aspect

# Load the image into an array: img
img = plt.imread('Astronaut.jpg')

# Specify the extent and aspect ratio of the top left subplot
plt.subplot(2,2,1)
plt.title('extent=(-1,1,-1,1),\naspect=0.5') 
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=0.5)

# Specify the extent and aspect ratio of the top right subplot
plt.subplot(2,2,2)
plt.title('extent=(-1,1,-1,1),\naspect=1')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=1)

# Specify the extent and aspect ratio of the bottom left subplot
plt.subplot(2,2,3)
plt.title('extent=(-1,1,-1,1),\naspect=2')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=2)

# Specify the extent and aspect ratio of the bottom right subplot
plt.subplot(2,2,4)
plt.title('extent=(-2,2,-1,1),\naspect=2')
plt.xticks([-2,-1,0,1,2])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-2,2,-1,1), aspect=2)

# Improve spacing and display the figure
plt.tight_layout()
plt.show()
````

````python
# Load the image into an array: image
image = plt.imread('Unequalized_Hawkes_Bay_NZ.jpg')

# Extract minimum and maximum values from the image: pmin, pmax
pmin, pmax = image.min(), image.max()
print("The smallest & largest pixel intensities are %d & %d." % (pmin, pmax))

# Rescale the pixels: rescaled_image
rescaled_image = 256*(image - pmin) / (pmax - pmin)
print("The rescaled smallest & largest pixel intensities are %.1f & %.1f." % 
      (rescaled_image.min(), rescaled_image.max()))

# Display the original image in the top subplot
plt.subplot(2,1,1)
plt.title('original image')
plt.axis('off')
plt.imshow(image)

# Display the rescaled image in the bottom subplot
plt.subplot(2,1,2)
plt.title('rescaled image')
plt.axis('off')
plt.imshow(rescaled_image)

plt.show()
````
