# Plotting multiple graphs


## Strategies

- plotting many graphs on common axes
- creating axes with afigure
- creating subplots withing a figure


```python
import csv
csvfile = open('percent-bachelors-degrees-women-usa.csv', 'r')
degrees_woman = csv.DictReader(csvfile)

#['Year', 'Agriculture', 'Architecture', 'Art and Performance', 'Biology', 'Business', 'Communications and Journalism', 'Computer Science', 'Education', 'Engineering', 'English', 'Foreign Languages', 'Health Professions', 'Math and Statistics', 'Physical Sciences', 'Psychology', 'Public Administration', 'Social Sciences and History']

year = []
physical_sciences = []
computer_science = []
health = []
education = []

for row in degrees_woman:
    year.append(int(row['Year']))
    physical_sciences.append(float(row['Physical Sciences']))
    computer_science.append(float(row['Computer Science']))
    health.append(float(row['Health Professions']))
    education.append(float(row['Education']))

year = np.array(year)
physical_sciences = np.array(physical_sciences)
computer_science = np.array(computer_science)
health = np.array(health)
education = np.array(education)
```

```python
## EXAMPLES 1.1

# Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Plot in blue the % of degrees awarded to women in the Physical Sciences
plt.plot(year,physical_sciences, color='blue')

# Plot in red the % of degrees awarded to women in Computer Science
plt.plot(year,computer_science, color='red')


# Display the plot
plt.show()
```

```python
## Example 1.2

# Create plot axes for the first line plot
plt.axes([0.05,0.05, 0.425, 0.9])

# Plot in blue the % of degrees awarded to women in the Physical Sciences
plt.plot(year, physical_sciences, color='blue')

# Create plot axes for the second line plot
plt.axes([0.525, 0.05, 0.425, 0.9])

# Plot in red the % of degrees awarded to women in Computer Science
plt.plot(year, computer_science, color='red')

# Display the plot
plt.show()
```

```python
# EXAMPLE 1.3


# Create a figure with 1x2 subplot and make the left subplot active
plt.subplot(1,2,1)

# Plot in blue the % of degrees awarded to women in the Physical Sciences
plt.plot(year, physical_sciences, color='blue')
plt.title('Physical Sciences')

# Make the right subplot active in the current 1x2 subplot grid
plt.subplot(1,2,2)

# Plot in red the % of degrees awarded to women in Computer Science
plt.plot(year, computer_science, color='red')
plt.title('Computer Science')

# Use plt.tight_layout() to improve the spacing between subplots
plt.tight_layout()
plt.show()
```

```python
## Example 1.4

# Create a figure with 2x2 subplot layout and make the top left subplot active
plt.subplot(2,2,1)

# Plot in blue the % of degrees awarded to women in the Physical Sciences
plt.plot(year, physical_sciences, color='blue')
plt.title('Physical Sciences')

# Make the top right subplot active in the current 2x2 subplot grid 
plt.subplot(2,2,2)

# Plot in red the % of degrees awarded to women in Computer Science
plt.plot(year, computer_science, color='red')
plt.title('Computer Science')

# Make the bottom left subplot active in the current 2x2 subplot grid
plt.subplot(2,2,3)

# Plot in green the % of degrees awarded to women in Health Professions
plt.plot(year, health, color='green')
plt.title('Health Professions')

# Make the bottom right subplot active in the current 2x2 subplot grid
plt.subplot(2,2,4)

# Plot in yellow the % of degrees awarded to women in Education
plt.plot(year, education, color='yellow')
plt.title('Education')

# Improve the spacing between subplots and display them
plt.tight_layout()
plt.show()
```

# Customizing axes


## Controlling axis extendts

- `axis([xmin, xmax, ymin, ymax])` sets axis extents
- Controll over individual axis extents
    - `xlim([xmin, xmax])`
    - `ylim([ymin, ymax])`
- CAn use tuples, list for extents
    - `xlim((-2,3))` works
    - `xlim([-2,3])` works also
    
    


```python
import csv
csvfile = open('gdp_csv.csv', 'r')
gdp = csv.DictReader(csvfile)
# keys ['Country Name', 'Country Code', 'Year', 'Value'])

yr =[]
us_gdp = []

for row in gdp:
    #country = row.pop('Country Code')
    #print(country)
    #print(row['Country Code'],row['Year'])
    if row['Country Code'] == 'USA':
        yr.append(int(row['Year']))
        us_gdp.append(int(row['Value'])/1000000000)

yr = np.array(yr)
us_gdp = np.array(us_gdp)
```

```python
us_gdp.max()
```

```python
import matplotlib.pyplot as plt
plt.plot(yr,us_gdp)
#plt.yticks(min(us_gdp),max(us_gdp))

plt.xlabel("Year")
plt.ylabel('Billions of dollars')

plt.title('US gross Domestic product')

plt.show()
```

## Using `xlim()`

Lets use `xlim` to only display data from 1960 to 1970

```python
plt.plot(yr, us_gdp)
plt.xlabel('Year')
plt.ylabel('Billions of Dollars')

# here we use xlim to display only data from 1960 - 1970
plt.xlim((1960, 1970))

plt.show()
```

## Using `xlim()` & `ylim()`



```python
plt.plot(yr, us_gdp)
plt.xlabel('Year')
plt.ylabel('Billions of Dollars')

# here we use xlim to display only data from 1960 - 1970
plt.xlim((1960, 1970))
plt.ylim(0,2500)

plt.show()
```

## Using `axis()`

```python
plt.plot(yr, us_gdp)
plt.xlabel('Year')
plt.ylabel('Billions of Dollars')
plt.title('US Gross Domestic product')


plt.axis((1960, 1970, 0, 2500))
plt.show()
```

## Other `axis()` options



|  Invocation |Result   |
|--               -|---|
|`axis('off')`     |turns off axis lines, labels   |
| `axis('equal')`  |equal scaling on x, y axes   |
| `axis('square)`  |forces square plot   |
| `axis('tight')`  |sets `xlim()`, `ylim()` to show all data   |




```python
## EXAMPLE 1

# Plot the % of degrees awarded to women in Computer Science and the Physical Sciences
plt.plot(year,computer_science, color='red') 
plt.plot(year, physical_sciences, color='blue')

# Add the axis labels
plt.xlabel('Year')
plt.ylabel('Degrees awarded to women (%)')

# Set the x-axis range
plt.xlim([1990, 2010])

# Set the y-axis range
plt.ylim([0,50])

# Add a title and display the plot
plt.title('Degrees awarded to women (1990-2010)\nComputer Science (red)\nPhysical Sciences (blue)')
plt.show()

# Save the image as 'xlim_and_ylim.png'
plt.savefig('xlim_and_ylim.png')

```

## Legends, annotations, and styles

```python
setosa_slen = []
setosa_swid = []
versicolor_slen = []
versicolor_swid = []
virginica_slen = []
virginica_swid = []


data = open('iris.csv.txt', 'r')
# ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])
# setosa,versicolor,virginica
petal = csv.DictReader(data)


for row in petal:    
    if row['species'] == 'setosa':
        setosa_slen.append(float(row['sepal_length']))
        setosa_swid.append(float(row['sepal_width']))
    elif row['species'] == 'versicolor':
        versicolor_slen.append(float(row['sepal_length']))
        versicolor_swid.append(float(row['sepal_width']))
    elif row['species'] == 'virginica':
        virginica_slen.append(float(row['sepal_length']))
        virginica_swid.append(float(row['sepal_width']))
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.axis((4.0,8.5, 1.5, 5.0))
plt.scatter(setosa_slen, setosa_swid,
           marker = 'o', color = 'red', label = 'setosa')


plt.scatter(versicolor_slen, versicolor_swid,
           marker = 'o', color = 'green', label = 'versicolor')

plt.scatter(virginica_slen, virginica_swid,
           marker = 'o', color = 'blue', label = 'virginica')


plt.legend(loc='upper right')
plt.title('Iris data')

plt.xlabel('Sepal lenght (cm)')
plt.ylabel('sepal width (cm)')


plt.show()
```

## plot annotations

- text labels and arows using `annotate()`
- flexible specification of coordinates
- keyword aarowprops: dic of arow properties
    - widdth
    - color
    - etc

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.axis((4.0,8.5, 1.5, 5.0))

plt.scatter(setosa_slen, setosa_swid,
           marker = 'o', color = 'red', label = 'setosa')


plt.scatter(versicolor_slen, versicolor_swid,
           marker = 'o', color = 'green', label = 'versicolor')

plt.scatter(virginica_slen, virginica_swid,
           marker = 'o', color = 'blue', label = 'virginica')


plt.legend(loc='upper right')
plt.title('Iris data')

plt.xlabel('Sepal lenght (cm)')
plt.ylabel('sepal width (cm)')



# using annotate() for text

plt.annotate('setosa',xy=(5.0, 3.5))
plt.annotate('virginica', xy=(7.25,3.5))
plt.annotate('versicolor', xy=(5.0, 2.0))

plt.show()
```

## Options for `annotate()`


|option      |description   |
|------------|--------------|
| s          |text of label   |
| xy         |coordinates to annotate   |
| xytext     |coordinates of label   |
| arrowprops |controls drawing of arows   |

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.axis((4.0,8.5, 1.5, 5.0))

plt.scatter(setosa_slen, setosa_swid,
           marker = 'o', color = 'red', label = 'setosa')


plt.scatter(versicolor_slen, versicolor_swid,
           marker = 'o', color = 'green', label = 'versicolor')

plt.scatter(virginica_slen, virginica_swid,
           marker = 'o', color = 'blue', label = 'virginica')


plt.legend(loc='upper right')
plt.title('Iris data')

plt.xlabel('Sepal lenght (cm)')
plt.ylabel('sepal width (cm)')



# using annotate() for arrows

plt.annotate('setosa',xy=(5.0, 3.5),
            xytext = (4.25, 4.0),
            arrowprops={'color': 'red'})

plt.annotate('virginica', xy=(7.25,3.5),
            xytext = (6.5, 4.0),
            arrowprops={'color': 'blue'})

plt.annotate('versicolor', xy=(5.0, 2.0),
            xytext=(5.5, 1.75),
            arrowprops={'color': 'green'})

plt.show()
```
