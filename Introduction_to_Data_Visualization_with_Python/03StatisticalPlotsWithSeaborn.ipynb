## Recap Pandas DataFrames
- labelled tabular data structure
- labels on rows: *index*
- columns are Pandas Series


- `total_bill`, `tip` are contnous
- `sex`,`day`,`time`   are categorical, sometimes called factors. variables that have a (finite)

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

## Linear regression plots

- fitting a straight line through a scatter plot of points in 2D
- tip vs total bill points
- narow band higlighting 95% confidence interval for this regression

```python
sns.lmplot(x='total_bill', y='tip', data=tips)

plt.show()
```

## Factors
- grouping factors (sameplot)
    - `hue='what-to-factor'`
- grouping factors (subplots)
    - `col='what-to-factor`
    - creates two identical graphs separated by 'what-to-factor

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='Set1')

plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')

plt.show()
```

## Resisual plots
**How much the data misses the regression line**
- dotted line is 0 residual



- Using `residplot()`

- Similar arguments as `lmplot()` but more flexible
- x,y can be *arrays or strings*
- data is DataFrame (optional)
- optional arguments(color) as in matplotlib

```python
sns.residplot(x='total_bill', y='tip', data=tips, color='indianred')

plt.show()
```

```python
## Example
# Simple linear regressions
# load data
import pandas as pd

auto = pd.read_csv('auto-mpg.csv')

# Import plotting modules
import matplotlib.pyplot as plt
import seaborn as sns

# Plot a linear regression between 'weight' and 'hp'
sns.lmplot(x='weight', y='hp', data=auto)

# Display the plot
plt.show()
```

```python
## Example
# Plotting residuals of a regression

# Import plotting modules
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a green residual plot of the regression between 'hp' and 'mpg'
sns.residplot(x='hp', y='mpg', data=auto, color='green')

# Display the plot
plt.show()
```

```python
## example
# Higher-order regressions
# Generate a scatter plot of 'weight' and 'mpg' using red circles
plt.scatter(auto['weight'], auto['mpg'], label='data', color='red', marker='o')

# Plot in blue a linear regression of order 1 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, color='blue', scatter=None, label='order 1')

# Plot in green a linear regression of order 2 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, color='green', scatter=None, order=2, label='order 2')

# Add a legend and display the plot
plt.legend(loc='upper right')
plt.show()




```

```python
## EXAMPLE
# Grouping linear regressions by hue
# Plot a linear regression between 'weight' and 'hp', with a hue of 'origin' and palette of 'Set1'
sns.lmplot(x='weight', y='hp',data=auto,hue='origin', palette='Set1')

# Display the plot
plt.show()
```

```python
# EXAMPLE
# Grouping linear regressions by row or column
# Plot linear regressions between 'weight' and 'hp' grouped row-wise by 'origin'
sns.lmplot(x='weight', y='hp',row='origin', data=auto)

# Display the plot
plt.show()
```

## Visualizing data
- Univariate -> "one variable"
- Visualization techiques for sampled univariate data
    - Strip plots
    - Swarm plots
    - Violing plots

## Strip plot
- using `stripplot()`

```python
sns.stripplot(y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

## Grouping with `stripplot()`
- grouped strip plot

```python
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

## Spreading out strip plots

```python
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=True)
plt.ylabel('tip ($)')

plt.show()
```

## Swarm plot

```python
# using swarmplot()

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

## More grouping

```python
# more grouping w/ swarmplot()

sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')

plt.show()
```

## Changing orientation

```python
sns.swarmplot(x='tip', y='day', data=tips, hue='sex', orient='h')
plt.xlabel('tip ($)')

plt.show()
```

## Violin plot

```python
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.tight_layout()
plt.show()
```

## combining plots

```python
sns.violinplot(x='day', y='tip',
               data=tips, inner=None,
              color='lightgray')
sns.stripplot(x='day', y='tip',
              data=tips, size=4,
             jitter=True)

plt.ylabel('tip ($)')
plt.show()
```

```python
##Example
# Constructing strip plots
# Make a strip plot of 'hp' grouped by 'cyl'
plt.subplot(2,1,1)
sns.stripplot(x='cyl', y='hp', data=auto)

# Make the strip plot again using jitter and a smaller point size
plt.subplot(2,1,2)
sns.stripplot(x='cyl', y='hp', data=auto, size=3, jitter=True)

# Display the plot
plt.show()
```

```python
##Example
# Constructing swarm plots
# Generate a swarm plot of 'hp' grouped horizontally by 'cyl'  
plt.subplot(2,1,1)
sns.swarmplot(x='cyl',y='hp',data=auto)

# Generate a swarm plot of 'hp' grouped vertically by 'cyl' with a hue of 'origin'
plt.subplot(2,1,2)
sns.swarmplot(x='hp',y='cyl',data=auto, hue='origin', orient='h')

# Display the plot
plt.show()
```

```python
##Example
# Constructing violin plots
# Generate a violin plot of 'hp' grouped horizontally by 'cyl'
plt.subplot(2,1,1)
sns.violinplot(x='cyl', y='hp', data=auto)

# Generate the same violin plot again with a color of 'lightgray' and without inner annotations
plt.subplot(2,1,2)
sns.violinplot(x='cyl', y='hp', inner=None,data=auto, color='lightgray')

# Overlay a strip plot on the violin plot
sns.stripplot(x='cyl',y='hp', data=auto,jitter=True, size=1.5)

# Display the plot
plt.show()
```

## Joint plot
- scatter plot of total bill and tip values within the axis
- above and to the right are histograms of the in

    - person : how two values are coorelated
    - pvaue : statistical significance of the difference of this value from zero


```python
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.style.use('seaborn')
plt.show()
```

## Joint plot using KDE

Kernal Density Estimation
- used to smooth the central scatter plott and the bounding univariant distributions
- produces smmoth contour plots 

- using `kde=True`


### Kind
- kind='scatter' uses a scatter plot of the data points
- kind='reg' uses a regression plot (default order 1)
- kind='resid' uses a residual plot
- kind='kde' uses a kernel density estimate of the joint distribution
- kind='hex' uses a hexbin plot of the joint distribution

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')

plt.show()
```

## Covariance heat map of tips data

- heatmap: is alsmost the same as a pseduocolor plot of a matrix with aditional functionality for pandas dataframes


### Covariance
- entries of a covariance matrix quantify how strongly two variables change together.

```python
covariance = tips.corr()
```

```python
print(covariance)

sns.heatmap(covariance)
plt.title('Covariance plot')

plt.show()
```
