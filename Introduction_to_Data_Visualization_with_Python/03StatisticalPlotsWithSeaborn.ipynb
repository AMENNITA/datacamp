# Seaborn


## Recap Pandas DataFrames
- labelled tabular data structure
- labels on rows: *index*
- columns are Pandas Series


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

## Linear regression plots

```python
sns.lmplot(x='total_bill', y='tip', data=tips)

plt.show()
```

## Factors
- grouping factors (sameplot)
    - `hue='what-to-factor'`
- grouping factors (subplots)
    - `col='what-to-factor`
    - creates two identical graphs separated by 'what-to-factor

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='Set1')

plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')

plt.show()
```

## Resisual plots
- Using `residplot()`

- Similar arguments as `lmplot()` but more flexible
- x,y can be *arrays or strings*
- data is DataFrame (optional)
- optional arguments(color) as in matplotlib

```python
sns.residplot(x='total_bill', y='tip', data=tips, color='indianred')

plt.show()
```

## Visualizing data
- Univariate -> "one variable"
- Visualization techiques for sampled univariate data
    - Strip plots
    - Swarm plots
    - Violing plots

## Strip plot
- using `stripplot()`

```python
sns.stripplot(y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

## Grouping with `stripplot()`
- grouped strip plot

```python
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

## Spreading out strip plots

```python
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=True)
plt.ylabel('tip ($)')

plt.show()
```

## Swarm plot

```python
# using swarmplot()

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

## More grouping

```python
# more grouping w/ swarmplot()

sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')

plt.show()
```

## Changing orientation

```python
sns.swarmplot(x='tip', y='day', data=tips, hue='sex', orient='h')
plt.xlabel('tip ($)')

plt.show()
```

## Violin plot

```python
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.tight_layout()
plt.show()
```

## combining plots

```python
sns.violinplot(x='day', y='tip',data=tips, inner=None,
              color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4,
             jitter=True)

plt.ylabel('tip ($)')
plt.show()
```

## Visualizing data

- Bivariate --> 'two variable'
- Multivariate --> 'multiple variables'
- Visualizing relationship in multibariate data
    - joiint plots
    - pair plots
    - heat maps
    

## Joint plot

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.style.use('seaborn')
plt.show()
```

## Joint plot using KDE
- using `kde=True`

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')

plt.show()
```

## Pair plot

```python
sns.pairplot(tips)
plt.show()
```

### `pairplot()` with hue

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

## Covariance heat map of tips data

```python
covariance = tips.corr()
```

```python
print(covariance)

sns.heatmap(covariance)
plt.title('Covariance plot')

plt.show()
```
