# Plotting subset of your time series data

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
df_subset = discoveries['1900':'1930']
ax = df_subset.plot(color='blue', fontsize=14)
plt.show()
```

# Adding markers

```python
ax.axvline(x='1910-01-01', color='red', linestyle='--')  # vertical line
ax.axhline(y=100, color='green', linestyle='--')  # horisontal 
```

# Using markers: the full code

```python
ax = discoveries.plot(color='blue')
ax.set_xlabel('Date')
ax.set_ylabel('Number of great discoveries')
ax.axvline('1939-01-01', color='red', linestyle='--')
ax.axhline(4, color='green', linestyle='--')
```

# Highlighting regions of interest

```python
ax = discoveries.plot(color='blue')
ax.set_xlabel('Date')
ax.set_ylabel('Number of great discoveries')
ax.axvspan('1914-01-01', '1918-01-01', color='red', alpha=0.3)
ax.axhspan(8, 6, color='green', alpha=0.3)
```

# Add vertical and horizontal markers


```python
# Plot your the discoveries time series
ax = discoveries.plot(color='blue', fontsize=6)

# Add a red vertical line
ax.axvline(x='1939-01-01', color='red', linestyle='--')

# Add a green horizontal line
ax.axhline(y=4, color='green', linestyle='--')

plt.show()
```

# Add shaded regions to your plot


```python
# Plot your the discoveries time series
ax = discoveries.plot(color='blue', fontsize=6)

# Add a vertical red shaded region
ax.axvspan('1900-01-01', '1915-01-01', color='red', alpha=0.3)

# Add a horizontal green shaded region
ax.axhspan(8, 6, color='green', alpha=0.3)

plt.show()
```
