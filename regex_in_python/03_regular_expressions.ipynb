# Repetitions

## quantifier


- `\d{1,3}`: digit appears **1 upto 3 times**
- `\d+` : digit should appear **1 or more times**
    - character applies to the end of char or digit

- `\d*`: digit appear **0 or more times***


- `\d?` : digt appears **zero times or one**

## Greedy v non-greedy

## greedy matching
- greedy: match as many chars a possible
- return the longest match
- backtracks when toom ahchars are match
- gives up chars one at time

````python
import re
re.match(r'\d+', '12345cada')
````

````python
re.match(r'.*hello', 'xhelloxxxxxxx')
````

## Non-greedy
- lazy: match as few chars as needed
- return the shortes match
- append `?` to geedy quant

````python
re.match(r'\d+?','12345bcada')
````

````python
re.match(r'.*?hello', 'xhelloxxxxxxx')
````
