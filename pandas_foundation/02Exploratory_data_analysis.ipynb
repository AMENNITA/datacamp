# Visual exploratory data analysis

## The iris data set
- famous dataset in pattern recognition
- 150 observations, 4 features each
    - Sepal length
    - sepal width
    - petal length
    - petal width

- 3 species: Setosa, versicolor virginica

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
# iris = pd.DataFrame(iris.data, columns = iris.feature_names)
iris = pd.read_csv('iris_d.csv')
````

````python
iris.info()
````

````python
iris.head()
````

## Lineplot

````python
iris.plot(x='SepalLength', y='SepalWidth')
plt.show()
````

## Scatterplot

````python
iris.plot(x='SepalLength', y='SepalWidth', kind='scatter')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')

plt.show()

````

## Boxplot

````python
iris.plot(y='SepalLength',kind='box')
plt.ylabel('sepal width (cm)')

plt.show()
````

## Histogram

````python
iris.plot(y='SepalLength',kind='hist')
plt.xlabel('sepal length (cm)')

plt.show()
````

## Histogram options

- bins(integer): number of intervals or bins
- range(tuple): extrema of bins (min,max)
- normed(boolean): whether to normailze to one
- cumulative(boolean) compute Cumulative Distribution Function (CDF)
- ... more Matplotib customizations

````python
iris.plot(y='SepalLength',kind='hist', bins=30, range=(4,8), normed=True)

plt.xlabel('sepal length  (cm)')

plt.show()
````

## Cumulative distribution


````python
iris.plot(y='SepalLength',kind='hist', 
          bins=30, range=(4,8),
          cumulative = True, normed=True)

plt.xlabel('sepal length  (cm)')
plt.title('Cumulative distruvution function (CDF)')
plt.show()
````

## Word of warning

- three different dataframes plot idioms
    - `iris.plot(kind='hist')`
    - `iris.lt.hist()`
    - `iris.hist()`
- syntax/results differ!
- Pandas API  still evolbing: check documentation
